```python
# the issue is, we have the genres saved in string format and we need integer indexing
# we've to convert it in a list
# we have a module in python for this particular thing

import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')

#we can see the difference in output of above code and this one, so we'll use this ast in our defined function
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(1) #corrected movies and genres
```

```python
# to correct cast, we'll alter the function a lil since cast is stored in a diff format, we'll make use of top 3 lead roles since cast m we have 100s of names
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3: #making sure 3 actors included
            L.append(i['name'])
            counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
movies['crew'][0] 
#now how to change this? we only need names of director and no other crew so let's make a function accordingly
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'][0]
```

```python
# to change overview, we'll do something like this
movies['overview']=movies['overview'].apply(lambda x:x.split()) #converting each string into list of words
```

```python
#cast m we're taking top 3 lead roles
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
movies.head(1)
```

```python
movies['genre']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])


# def collapse(L):
#     L1 = []
#     for i in L:
#         L1.append(i.replace(" ",""))
#     return L1
# the function that we applied is something similar to this
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
#converting the list back to string
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))

#it's recommending loc instead of our method of changing which might cause error in near future
# The .loc indexer allows you to modify the DataFrame explicitly and helps in avoiding potential ambiguity that can trigger the 
# warning.

# In essence, the warning is suggesting that using .loc is a safer way to modify DataFrame values and helps prevent unintended 
# side effects due to chained indexing.
# It's a good practice to follow in order to write code that is less error-prone and easier to understand.
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower()) #converting into lowercase
```

```python
#we have certain features which are unnecessary for ex: action,actions (both are same and causing repitition)
#for this thing we'll apply stemming on the db new_db itself 
```

```python
import nltk #library used to do stemming, avoiding unnecessary words to get into vector (romance, romantic, romaticism- romanc)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#helper function to perform stemming
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
#to check how the stem function works, we applied it on a screen and this is the result
stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy Science Fiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d SamWorthington ZoeSaldana SigourneyWeaver JamesCameron')
```

```python
#now finally, we'll apply this stem function to our tags
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#we're importing this to implement vectorisation
# The CountVectorizer is a text feature extraction method that converts a collection of text documents into a matrix of token counts.
from sklearn.feature_extraction.text import CountVectorizer
# Initializing the CountVectorizer Object:
cv = CountVectorizer(max_features=5000,stop_words='english') 


#these hyperparamteres are present in sklearn
#maxwords is max words we want to include, stop words will exclude english fillers like is in the etc
```

```python
# The line vector = cv.fit_transform(new['tags']).toarray() is using the previously created CountVectorizer object (cv) to 
# transform the text data in the 'tags' column of the DataFrame new into a numeric matrix representation.
# refer to notes, you'll understand formation of table
vector = cv.fit_transform(new_df['tags']).toarray() #tags converted to array
```

```python
# we have 4806 movies which means 4806 vectors

# we have to calculate distance of each movie with other, we'll use sklearn.metrics to do this
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0] #pehli movie ka har movie ke sath score
#ek movie ki usse khudse similarity will be 1
```

```python
#now we've to define a function jisme ek text input arha h, uske corresponding movie_id find krenge
#movie id will be the index of vector, wo fetch karenge, pehle us particular vector ko enumerate krenge 
# for ex [10,76,56] becomes [(1,10),(2,76),(3,56)] 1d arrahy 2d bnjaega indices ke sath
```

```python
sorted_enum = sorted(enumerate(similarity[0]), reverse=True, key=lambda x: x[1]) #just to show how enumerate works, we did it in accordance with indices

sorted_enum
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]; #finding index of the recommended movie
    distances=similarity[movie_index] #similarity vector agya, jissme ratios h baki movies ki similarity ki
    movies_list= sorted(enumerate(distances), reverse=True, key=lambda x: x[1]) [1:6] #we want first 6 elements
    
    for i in movies_list:
        #i[0] gives us movie ka index jo chahie, usse uska title db se dhundlo
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend('Batman Begins')
```
